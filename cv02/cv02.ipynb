# Dataset preparation and visualization

First we prepare the dataset and visualize it. For straightforward visualization we can use pandas to store the data and
seaborn to plot the points.

```python
import os
import pandas as pd
import seaborn as sns

# Path to the dataset, downloaded from https://gist.github.com/netj/8836201
dataset_path = 'dataset/iris.csv'
```

```python
# Create pandas dataframe (i.e. a "table" that holds all the data)
df = pd.read_csv(dataset_path)
df
```

```python
# Here we see that we have 4 feature dimensions - sepal.length, sepal.width, petal.length, petal.width
# and a label - Setosa, Versicolor, Virginica

feature_cols = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
classes = ['Setosa', 'Versicolor', 'Virginica']
```

```python
# We have even distribution of classes
df['variety'].value_counts()
```

Lets project points to 2D space and visualize some of the dimensions

```python
sns.jointplot(x='sepal.length', y='sepal.width', hue='variety', data=df)
```

```python
sns.jointplot(x='petal.length', y='petal.width', hue='variety', data=df)
```

```python
sns.jointplot(x='sepal.width', y='petal.width', hue='variety', data=df)
```

```python
sns.jointplot(x='sepal.length', y='petal.length', hue='variety', data=df)
```
